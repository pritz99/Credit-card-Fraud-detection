```python
from operator import xor


# Separate the data by class
class_0 = df1[df1['Class'] == 0]
class_1 = df1[df1['Class'] == 1]

# Downsample the majority class
class_0_downsampled = class_0.sample(n=len(class_1)*3, random_state=42)

# Concatenate the downsampled majority class with the minority class
balanced_data = pd.concat([class_0_downsampled, class_1])

# Shuffle the balanced dataset
balanced_data = balanced_data.sample(frac=1, random_state=42).reset_index(drop=True)


x=balanced_data.drop('Class',axis=1)
y=balanced_data['Class']
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,classification_report,f1_score
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
scaler = MinMaxScaler()

# Fit the scaler on the training set and transform it
X_train_scaled = scaler.fit_transform(X_train)

# Transform the test set using the already-fitted scaler
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
KNN=KNeighborsClassifier(n_neighbors=5)
```

```python
KNN.fit(X_train_scaled,y_train)
```

```python
y_pred=KNN.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
print(accuracy_score(y_pred,y_test))
```

By downsampling, we achieve a higher and more improved F1 score compared to the previous results when running KNN on an imbalanced dataset. In this approach, instead of using the entire majority class, we limit its size to only one-third of the minority class.

```python
f1 = f1_score(y_test, y_pred)

# Print the F1 score
print("F1 Score:", f1)
```

When we apply stratified cross-validation, our F1 score improves by approximately 4-5%.

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
```

```python
skf=StratifiedKFold(n_splits=5, shuffle=True, random_state=42) 
```

```python
cv_scores = cross_val_score(KNN, X_train_scaled, y_train, cv=skf, scoring='f1')
print("Cross-validation scores:", cv_scores)
print("Mean CV accuracy:", cv_scores.mean())
```
